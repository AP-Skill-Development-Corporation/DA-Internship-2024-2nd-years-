### Line plot
* plt.plot()

```python
x_axis=[10,20,30,40,50]
y_axis=[1,2,3,4,5]
plt.plot(x_axis,y_axis,label='Line')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Line Graph')
plt.legend(loc='lower right')
```

```python
help(plt.legend)
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,20)
x
```

```python
plt.plot(x,x**2,color='red',label='Line1')
plt.plot(x,x**3,color='green',label='Line2')
```
