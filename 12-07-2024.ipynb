```python
import matplotlib.pyplot as plt
```

### Horizontal Bar Graph
* plt.barh()

```python
import numpy as np
x=np.arange(5)
x
```

```python
plt.barh(x,[90,80,50,60,70],color='r')
```

```python
x=[1,2,3,4,5]
y=[10,55,23,44,67]
plt.barh(x,y,color=['m','gray','green','pink','r'])
```

### Scatter plot
* It is similar to Line plot
* plt.scatter()

```python
plt.scatter([1,3],[5,6])
```

```python
x=np.random.randint(100,1000,100)
y=np.random.randint(500,1500,100)
```

```python
plt.scatter(x,y)
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
```

plt.scatter(x,y,color='red',marker='|')

### Histogram

* Histogram is used to show the frequency distribution
* plt.hist()



```python
a=[0,0.1,0.9,2,2.2,2.2,3,5,5,5.6]
bins=4
plt.hist(a,bins,width=0.4)
```

```python
### pie chart
games=['cricket','chess','ludo']
scores=[90,78,65]
plt.pie(scores,labels=games,colors=['c','b','m'])
plt.show()
```
